```python
import pandas as pd

result=pd.read_csv("D:\Dataset\play_dataset.csv")
result.dropna()
```

```python
name_1_result=[]
task=['A','B','C']
result_1=result[result['name']=='name_1']
for i in task:
    save=[]
    result_1_A=result_1[result['task']==i]
    for j in range(1,16):
        save.append(result_1_A[result['channel']==j]['avg'].values[0])
    name_1_result.append(save)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)

plt.figure(figsize=(20, 30))

for i in range(1,16):
    plt.subplot(5,3,i)
    plt.title('name_1 Bar plot of channel'+str(i)+" while play")
    task = ['A', 'B', 'C']
    values = [name_1_result[0][i-1], name_1_result[1][i-1],name_1_result[2][i-1]]
    plt.bar(x, values)
    plt.xticks(x, task)
```
