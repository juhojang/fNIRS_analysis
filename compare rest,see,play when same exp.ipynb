# 실험(A,B,C) rest 구간 채널별로 저장

```python
import pandas as pd

result_rest=pd.read_csv("D:/Dataset/f_Nirs datasets/first_rest_dataset.csv")
result_rest.fillna(result_rest.mean(),inplace=True)

dict_rest={}

for l in list(set(result_rest['name'])):
    name_1_result=[]
    task=['A','B','C']
    result_1=result_rest[result_rest['name']==l]
    for i in task:
        save=[]
        result_1_A=result_1[result_rest['task']==i]
        for j in range(1,16):
            save.append(result_1_A[result_rest['channel']==j]['avg'].values[0])
        name_1_result.append(save)
    dict_rest[l]=name_1_result
```

# 실험(A,B,C) see 구간 채널별로 저장

# rest 구간, play 구간 비교

## task A

```python
import matplotlib.pyplot as plt
import numpy as np

task_list=['A_rest','A_play']
plt.figure(figsize=(20, 30))
plt.title(" 1~15")
for i in range(1,16):
    plt.subplot(5,3,i)
    plt.boxplot([A_rest[i-1],A_play[i-1]])
    plt.xticks([1,2], task_list)
    plt.title('CH '+str(i)+' box plot of A_rest, A_play')
plt.show()

import scipy.stats
for i in range(1,16):
    print('\n')
    print("CH "+str(i)+" paired t-test result")
    print(scipy.stats.ttest_rel(A_rest[i-1],A_play[i-1])) 
    #동일 피실험자로 간주하여 대응 표본 t 검정을 수행
    #귀무가설 : 두개의 데이터에는 유의미한 차이가 없을 것이다.
    #대립가설 : 두개의 데이터에는 유의미한 차이가 있을 것이다.
    #pvalue가 0.05(threshold) 이하일 경우에는 귀무가설을 기각하고 대립가설을 받아들임.
```
