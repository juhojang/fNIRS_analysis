```python
import pandas as pd
import numpy as np

play_blocks=pd.read_csv("D:\Dataset\play_dataset.csv")
play_blocks.dropna()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 150))
plt.title(" 1~15")
for i in range(1,16):
    plt.subplot(15,1,i)
    plt.title('Scatter plot of channel'+str(i)+" while play")
    channel=play_blocks[play_blocks['channel']==i]
    groups=channel.groupby('task')
    for name,group in groups:
        plt.scatter(group['slope'],group['avg'])
        plt.xlabel("slope")
        plt.ylabel("avg")
        plt.legend(["expA","expB","expC"],loc="center left")
```

```python
plt.figure(figsize=(20, 150))
plt.title(" 1~15")
for i in range(1,16):
    plt.subplot(15,1,i)
    plt.title('Scatter plot of channel'+str(i)+" while play")
    channel=play_blocks[play_blocks['channel']==i]
    groups=channel.groupby('task')
    for name,group in groups:
        plt.scatter(group['slope'],group['variance'])
        plt.xlabel("slope")
        plt.ylabel("variance")
        plt.legend(["expA","expB","expC"],loc="center left")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 150))
for i in range(1,16):
    channel=play_blocks[play_blocks['channel']==i]
    groups=channel.groupby('task')
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(projection='3d')
    for name,group in groups:    
        ax.scatter(group['slope'],group['avg'], group['variance'], marker='o', s=15, cmap='Greens')
        ax.set_xlabel("slope")
        ax.set_ylabel("avg")
        ax.set_zlabel("variance")
```

```python
plt.figure(figsize=(20, 30))
plt.title("Scatter plot of channel 1~15 while play")
task=['A','B','C']
for i in range(0,3):
    plt.subplot(3,1,i+1)
    tasks=play_blocks[play_blocks['task']==task[i]]
    groups=tasks.groupby('channel')
    for name,group in groups:
        plt.scatter(group['avg'],group['variance'])
```
