# 실험(A,B,C) rest 구간 채널별로 저장

```python
import pandas as pd

result_rest=pd.read_csv("first_rest_dataset.csv")
result_rest.fillna(result_rest.mean(),inplace=True)

dict_rest={}

for l in list(set(result_rest['name'])):
    name_1_result=[]
    task=['A','B','C']
    result_1=result_rest[result_rest['name']==l]
    for i in task:
        save=[]
        result_1_A=result_1[result_rest['task']==i]
        for j in range(1,16):
            save.append(result_1_A[result_rest['channel']==j]['avg'].values[0])
        name_1_result.append(save)
    dict_rest[l]=name_1_result
```

```python
A_rest=[]
B_rest=[]
C_rest=[]
for i in range(1,16):
    save_A=[]
    save_B=[]
    save_C=[]
    for j in list(dict_rest.keys()):
        save_A.append(dict_rest[j][0][i-1])
        save_B.append(dict_rest[j][1][i-1])
        save_C.append(dict_rest[j][2][i-1])
    A_rest.append(save_A)
    B_rest.append(save_B)
    C_rest.append(save_C)
```

# rest-play 정규성 검증

```python
import numpy as np
from scipy.stats import shapiro

A_rp=[]
B_rp=[]
C_rp=[]

for i in range(0,15):
    save_rp_A=[]
    save_rp_B=[]
    save_rp_C=[]
    for j in list(dict_see.keys()):
        save_rp_A.append(dict_rest[j][0][i]-dict_play[j][0][i])
        save_rp_B.append(dict_rest[j][1][i]-dict_play[j][1][i])
        save_rp_C.append(dict_rest[j][2][i]-dict_play[j][2][i])
    A_rp.append(save_rp_A)
    B_rp.append(save_rp_B)
    C_rp.append(save_rp_C)
```

```python
for i in range(1,16):
    print('\n')
    print("rest-play CH "+str(i)+" paired t-test result")
    shapiro_test = shapiro(A_rp[i-1])
    print(shapiro_test)
    #p_value가 0.05 이상일 경우 정규분포를 따름.
```
