# 1일차 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
from tslearn.metrics import dtw
from tslearn.clustering import TimeSeriesKMeans
from tslearn.utils import to_time_series_dataset

fNirs_list=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
score_list=[]
for i in range(1,7):
    temp=[]
    raw_fNirs=pd.read_csv("D:/Dataset/f_Nirs datasets/raw_data/Task A/A_K-"+str(i)+"_raw.csv")
    for j in range(0,15):
        fNirs_list[j].append(raw_fNirs.iloc[:,j])
```

```python
from tslearn.clustering import silhouette_score
total_label=[[],[],[],[],[],[]]
for i in range(0,15):
    X=to_time_series_dataset(fNirs_list[i])
    model = TimeSeriesKMeans(n_clusters=3, metric="dtw",
                            max_iter=10, random_state=0)
    model.fit(X)
    label=model.predict(X)
    total_label[0].append(label[0])
    total_label[1].append(label[1])
    total_label[2].append(label[2])
    total_label[3].append(label[3])
    total_label[4].append(label[4])
    total_label[5].append(label[5])
    print(silhouette_score(X,label,metric='dtw'))
```

```python
total_label
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
kmeans=KMeans(n_clusters=2,random_state=0)
clusters=kmeans.fit(total_label)
print(silhouette_score(total_label,clusters.labels_))
clusters.labels_
```

```python
self_report=pd.read_csv("D:/Dataset/f_Nirs datasets/k1~12_self-report_dataset.csv")
day1=self_report[0:6]
day1.drop(['subjectkey','date','age','sex','right_handed','piano_year','piano_major','impro_experience','condition','post_C_IF_10','pre_SE_10','pre_SE_11','post_B_IF_10','post_B_diff','post_C_diff'],axis=1,inplace=True)
for i in range(1,10):
    day1.drop(['pre_SE_0'+str(i)],axis=1,inplace=True)
    day1.drop(['post_C_IF_0'+str(i)],axis=1,inplace=True)
    day1.drop(['post_B_IF_0'+str(i)],axis=1,inplace=True)
day1
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
day1_pca = pca.fit_transform(day1)
day1_pcaDf = pd.DataFrame(data=day1_pca, columns = ['principal component1', 'principal component2'])
day1_pcaDf
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
kmeans=KMeans(n_clusters=2,random_state=0)
clusters=kmeans.fit(day1_pcaDf)
print(silhouette_score(day1_pcaDf,clusters.labels_))
clusters.labels_
```
