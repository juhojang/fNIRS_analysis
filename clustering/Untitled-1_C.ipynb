# 1일차 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
from tslearn.metrics import dtw
from tslearn.clustering import TimeSeriesKMeans
from tslearn.utils import to_time_series_dataset

fNirs_list=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
score_list=[]
for i in range(1,7):
    temp=[]
    raw_fNirs=pd.read_csv("D:/Dataset/f_Nirs datasets/raw_data/Task C/C_K-"+str(i)+"_raw.csv")
    for j in range(0,15):
        fNirs_list[j].append(raw_fNirs.iloc[:,j])
```

```python
from tslearn.clustering import silhouette_score
total_label=[[],[],[],[],[],[]]
for i in range(0,15):
    X=to_time_series_dataset(fNirs_list[i])
    model = TimeSeriesKMeans(n_clusters=2, metric="dtw",
                            max_iter=10, random_state=0)
    model.fit(X)
    label=model.predict(X)
    total_label[0].append(label[0])
    total_label[1].append(label[1])
    total_label[2].append(label[2])
    total_label[3].append(label[3])
    total_label[4].append(label[4])
    total_label[5].append(label[5])
    print(silhouette_score(X,label,metric='dtw'))
```
